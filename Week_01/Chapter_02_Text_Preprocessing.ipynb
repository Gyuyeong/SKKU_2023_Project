## Tokenization

주어진 corpus에서 token이라 불리는 단위로 나누는 작업

Token으로 나누는 방식은 tokenizer마다 조금씩 차이가 있을 수 있음.

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import WordPunctTokenizer
from tensorflow.keras.preprocessing.text import text_to_word_sequence
```

영어에서 Don't와 같이 apostrophe가 들어간 단어를 tokenize하는 방법은 다양함.
각 tokenizer들이 해당 경우를 어떻게 tokenize하는지 살펴보기

```python
print('word_tokize :', word_tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery  goes for a pastry shop."))
```

```python
print("WordPunctTokenizer :", WordPunctTokenizer().tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery  goes for a pastry shop."))
```

```python
print("text_to_word_sequence :", text_to_word_sequence("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery  goes for a pastry shop."))
```

각 tokenizer마다 "don't" 라는 단어를 tokenize하는 방법이 모두 다른 걸 확인할 수 있음.

word_tokenize는 "Don't"를 "Do", "n't"로 쪼갰고, WordPunctTokenizer는 "Don", "'", "t"로 쪼갰고, text_word_sequence는 "don't"로 그대로 보존하되, 모든 글자를 소문자로 바꿨음.

### 고려 사항

1. Punctuation을 전부 없애는 건 안된다. Ph.d 와 같은 단어는 "."이 꼭 필요하다. 
2. Apostrophe는 줄임말의 역할을 하는 경우가 있으므로, 이를 고려해야 함.
3. 공백을 함부로 다 없애면 안됨. 공백을 포함해서 한 단어가 되는 경우가 존재함. ex) Rock 'n' Roll

각 tokenizer는 저마다 규칙을 가지고 있음.

아래의 예시는 Penn Treebank Tokenizer이고 다음과 같은 규칙을 따라서 tokenize한다

1. hyphen(-)으로 구성된 단어는 하나로 유지
2. apostrophe로 접어가 함께하는 경우는 분리시킨다. ex) doesn't => "does" "n't"

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

text = "Starting a home-based restaurant may be an ideal. it doesn't have a food chain or restaurant of their own."
print(tokenizer.tokenize(text))
```

## Cleaning and Normalization

Cleaning: corpus의 noise를 제거하는 과정. tokenize 전후에 계속 실시해야 하는 것이지만, 완벽하게 정제하는 것은 거의 힘들다. 그래서 보통 합의점을 찾음

Normalization: 같은 단어인데 표현이 다른 경우, 이 단어들을 통합시켜 같은 단어들로 만들어 주는 단계

Normalization
1. 직접 규칙을 정해주는 경우: 예시, US와 USA는 같은 단어이니까 통합한다
2. 대, 소문자 관련: 기본적으로 대문자와 소문자의 차이로 인해 같은 단어가 다르게 tokenize 된 경우, 통합해줄 필요가 있음. 문장의 첫 글자는 항상 대문자여서 발생하는 현상이 잦음. 
**주의** : 대문자 소문자를 반드시 구별해야 하는 경우도 생김 ex) us(우리) US(미국) 

Cleaning: 불필요한 단어들을 제거하는 단계, 불필요하다는 것은 아무 의미 없는 것들 뿐만 아니라 목적에 맞지 않아서 불필요한 경우도 제거함을 뜻함

1. 등장 빈도가 너무 적은 단어
2. 길이가 짧은 단어 (영어의 경우, 단어의 길이가 평균적으로 6개 정도 되기 때문에 길이가 2~3 인 단어들을 버리는 것을 고려해볼 수 있다.) 주의: dog, cat은 3글자이다.

```python
import re
text = "I was wondering if anyone out there could enlighten me on this car."

shortword = re.compile(r'W*\b\w{1,2}\b')
print(shortword.sub('', text))
```

Regular Expression을 활용하면, 규칙적으로 나오는 패턴의 문자열들을 제거하기 용이함.

## 표제어 추출 (Lemmatization)

표제어: 기본 사전형 단어, 즉, 사전에 등재된 단어의 원형

정규화 기법 중 하나로 token에서 표제어를 추출하는 방법 ex) am, are, is => be

### 형태소

형태소: 의미를 가진 가장 작은 단위

- 어간(Stem): 단어의 의미를 담고 있는 부분
- 접사(affix): 단어에 추가적인 의마를 주는 부분

형태학적 parsing: 어간과 접사를 분리하는 작업 ex) cats => "cat" + "s"

- 자립(독립)형태소: 자체적으로 의미를 가지며 다른 형태소들과의 결합없이도 사용 가능한 형태소 ex) 명사 (의존명사는 예외)
- 의존 형태소: 다른 형태소와의 결합을 통해서만 사용할 수 있는 형태소 ex) 동사의 어간, 조사

NLTK에서 표제어 추출 방법

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ['policy', 'organization', 'have', 'going', 'love', 'lives', 'fly', 'dies', 'watched', 'has', 'starting']

print('표제어 추출 전 :', words)
print('표제어 추출 후 :', [lemmatizer.lemmatize(word) for word in words])
```

몇몇 알 수 없는 단어들은 품사를 같이 알려줘야 한다.

```python
lemmatizer.lemmatize('dies', 'v')
```

```python
lemmatizer.lemmatize('watched', 'v')
```

```python
lemmatizer.lemmatize('has', 'v')
```

### 어간 추출 (Stemming)

어간 추출은 정해진 규칙을 이용해서 parsing을 진행하기에 결과물을 보면 사전에 없는 단어들이 나타날 수 있다.

일반적으로 표제어 추출보다 빠른 속도로 이루어질 수 있음

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

sentence = "This was not the map we found in Billy Bones's chest, but an accurate copy, complete in all things--names and heights and soundings--with the single exception of the red crosses and the written notes."
tokenized_sentence = word_tokenize(sentence)

print('어간 추출 전 :', tokenized_sentence)
print('어간 추출 후 :', [stemmer.stem(word) for word in tokenized_sentence]) 
```

Porter 알고리즘은 기본적으로 "-ize", "-ance", "-al" 등의 어미를 삭제하는 규칙을 사용

```python
words = ['formalize', 'allowance', 'electrical']

print([stemmer.stem(word) for word in words])
```

Porter Stemmer와 Lancaster Stemmer 비교

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer

porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()

words = ['policy', 'doing', 'organization', 'have', 'going', 'love', 'lives', 'fly', 'dies', 'watched', 'has', 'starting']
print('어간 추출 전 :', words)
print('포터 스테머의 어간 추출 후:',[porter_stemmer.stem(w) for w in words])
print('랭커스터 스테머의 어간 추출 후:',[lancaster_stemmer.stem(w) for w in words])
```

자신의 상황에 맞는 어간 추출기를 사용해야 한다. 위 랭커스터 스테머의 경우, organization => organ으로 추출했는데, 이는 잘못되었을 가능성이 높다.

### Stopword

분석할 때 그닥 필요없다고 판단되는 단어 token들을 말함. NLTK에 약 100여개 이상의 불용어들이 정의되어 있음음

```python
from nltk.corpus import stopwords

stop_words_list = stopwords.words('english')
print('불용어 개수 :', len(stop_words_list))
print('불용어 10개 출력 :',stop_words_list[:10])
```

Stopword 제거하기

```python
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english'))

word_tokens = word_tokenize(example)

result = []

for word in word_tokens:
    if word not in stop_words:
        result.append(word)

print('불용어 제거 전 :',word_tokens) 
print('불용어 제거 후 :',result)
```

한글도 불용어 제거 가능. 사용자가 직접 정의해서 불용어를 제거해보기기

```python
okt = Okt()

example = "고기를 아무렇게나 구우려고 하면 안 돼. 고기라고 다 같은 게 아니거든. 예컨데 삼겹살을 구울 때는 중요한 게 있지."
stop_words = "를 아무렇게나 구 우려 고 안 돼 같은 게 구울 때 는"

stop_words = set(stop_words.split(' '))
word_tokens = okt.morphs(example)

result = [word for word in word_tokens if not word in stop_words]

print('불용어 제거 전 :',word_tokens) 
print('불용어 제거 후 :',result)
```

```python
r = re.compile("ab.")

r.match("kkkabc")
```

## Integer Encoding

```python
raw_text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
```

```python
sentences = sent_tokenize(raw_text)
print(sentences)
```

```python
vocab = {}
preprocessed_sentences = []
stop_words = set(stopwords.words('english'))

for sentence in sentences:
    tokenized_sentence = word_tokenize(sentence)
    result = []

    for word in tokenized_sentence:
        word = word.lower()
        if word not in stop_words:  # stopword가 아니어야 한다
            if len(word) > 2:  # 길이가 3 이상인 단어만 받는다
                result.append(word)
                if word not in vocab:
                    vocab[word] = 0
                vocab[word] += 1
    preprocessed_sentences.append(result)
print(preprocessed_sentences)
```

```python
all_words_list = sum(preprocessed_sentences, [])
print(all_words_list)
```

```python
from collections import Counter
```

```python
vocab = Counter(all_words_list)
print(vocab)
```

```python
vocab_size = 5
vocab = vocab.most_common(vocab_size) # 상위 5개
vocab
```

```python
word_to_index = {}
# 빈도수가 높을수록 더 낮은 index 부여
i = 0
for (word, frequency) in vocab:
    i = i + 1
    word_to_index[word] = i

print(word_to_index)
```

## Padding

자연어는 길이가 제각각이지만, 기계는 반드시 고정된 길이의 데이터를 요구하기에, 모든 데이터들의 길이를 맞춰주는 작업을 padding이라고 함

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
preprocessed_sentences = [['barber', 'person'], ['barber', 'good', 'person'], ['barber', 'huge', 'person'], ['knew', 'secret'], ['secret', 'kept', 'huge', 'secret'], ['huge', 'secret'], ['barber', 'kept', 'word'], ['barber', 'kept', 'word'], ['barber', 'kept', 'secret'], ['keeping', 'keeping', 'huge', 'secret', 'driving', 'barber', 'crazy'], ['barber', 'went', 'huge', 'mountain']]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(preprocessed_sentences)
encoded = tokenizer.texts_to_sequences(preprocessed_sentences)
print(encoded)
```

```python
max_len = max(len(item) for item in encoded)
print('최대 길이 :', max_len)
```

최대 길이에 맞춰 다른 문장들도 0으로 채워준다. 이때 0으로 채워주면 zero padding이라고 한다.

```python
for sentence in encoded:
    while len(sentence) < max_len:
        sentence.append(0)

padded_np = np.array(encoded)
padded_np
```

### Padding with Keras

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

encoded = tokenizer.texts_to_sequences(preprocessed_sentences)
print(encoded)
```

```python
padded = pad_sequences(encoded)
padded
```

텍스트 뒤를 padd하고 싶다면 `padding=post`를 추가한다.

```python
padded = pad_sequences(encoded, padding='post')
padded
```

Padding 할 때, 최대 길이에 맞추는 대신, 특정 길이로 제한하는 것도 가능함함. 

대신 이렇게 하면, 몇몇 sentence들은 일부가 잘려나갈 수 있음.

```python
padded = pad_sequences(encoded, padding='post', maxlen=5)
padded
```

잘리는 sentence들 중, 뒤를 잘리게 하고 싶다면 `truncating='post'`를 준다

```python
padded = pad_sequences(encoded, padding='post', maxlen=5, truncating='post')
```

```python
padded
```

때로는 다른 숫자로 padding을 하고 싶을 수도 있으니 그때는 `value`를 적어준다.

```python
padded = pad_sequences(encoded, padding='post', maxlen=5, value=14)
padded
```

## Vocabulary 단어 집합

단어 집합은 중복을 허용하지 않은 서로 다른 단어들의 집합이다. 이 단어 집합에 고유한 정수를 부여하면 encoding이 완료된다.

## One-Hot Encoding

앞서 단어마다 고유한 정수를 부여했는데, one-hot encoding은 고유한 정수가 아닌 고유한 벡터를 부여한다.전부 0인 벡터에 해당 단어의 고유한 정수에 해당하는 위치에만 1을 넣은 형태의 벡터가 완성된다. 이러한 벡터를 One-Hot Vector라고 함.

```python
from konlpy.tag import Okt

okt = Okt()
tokens = okt.morphs("나는 자연어 처리를 배운다")
print(tokens)
```

```python
word_to_index = {word : index for index, word in enumerate(tokens)}
print('단어 집합 :', word_to_index)
```

```python
def one_hot_encoding(word, word_to_index):
    one_hot_vector = [0]*(len(word_to_index))
    index = word_to_index[word]
    one_hot_vector[index] = 1
    return one_hot_vector

one_hot_encoding("자연어", word_to_index)
```

자연어는 index 값이 2 이므로, index 2의 위치에만 1이 저장되어 있는 걸 볼 수 있다.

### One-Hot Encoding in Keras

```python
text = "나랑 점심 먹으러 갈래 점심 메뉴는 햄버거 갈래 갈래 햄버거 최고야"
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
print('단어 집합 :', tokenizer.word_index)
```

```python
sub_text = "점심 먹으러 갈래 메뉴는 햄버거 최고야"
encoded = tokenizer.texts_to_sequences([sub_text])[0]
print(encoded)
```

```python
one_hot = to_categorical(encoded)
print(one_hot)
```

One-Hot encoding의 단점:
1. 각 단어의 크기가 너무 커진다
2. 단어들 간의 유사도를 나타낼 방법이 없다.
